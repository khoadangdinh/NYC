# NYC Population

Some key findings:
1. Staten Island is the only borough with consistent population growth since the 50s.
2. NYC Population hasn't changed that much.
3. The Bronx and Brooklyn have similar population changes over the years.
   

```python
import numpy as np
import matplotlib 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path='/Users/khoadinh/Desktop/IBM_Data_Science_Course/nycPop.csv'
df = pd.read_csv(path,low_memory=False)
#df.head()
df=df[['Borough','1950','1960','1970','1980'
       ,'1990','2000','2010','2020']] #only some columns
totalPop=df[0:1]
df=df[1:]#drop first row
df=df.T
#df.head()
df.columns=df.iloc[0]
df=df[1:]
df.head()

#
```

```python
categories=df.index
values1=df['   Bronx']
values2=df['   Brooklyn']
values3=df['   Manhattan']
values4=df['   Queens']
values5=df['   Staten Island']
```

```python

# Plot stacked bars
plt.figure(figsize=(6, 4), dpi=600)
plt.bar(categories, values1,color='royalblue', 
         label='Bronx')# first layer
plt.bar(categories, values2, bottom=values1,color='olive', 
         label='Brooklyn')
plt.bar(categories, values3, bottom=values2+values1,color='gray', 
         label='Manhattan')
plt.bar(categories, values4, bottom=values3+values2+values1,
         color='orange',label='Queens')
plt.bar(categories, values5, bottom=values4+values3+values2+values1,
         color='tomato',label='Staten Island')

# Add legend, labels, and title
plt.legend()
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population in Five Boroughs of NYC')
plt.legend(bbox_to_anchor=(1.02, 0.8), loc='upper left', borderaxespad=0)
# Show plot
plt.savefig("nyc_pop.jpg",bbox_inches="tight")
plt.show()
```

```python

# Plot stacked bars
plt.figure(figsize=(6, 4), dpi=600)
plt.barh(categories, values1,color='royalblue', 
         label='Bronx')# first layer
plt.barh(categories, values2, left=values1,color='olive', 
         label='Brooklyn')
plt.barh(categories, values3, left=values2+values1,color='gray', 
         label='Manhattan')
plt.barh(categories, values4, left=values3+values2+values1,
         color='orange',label='Queens')
plt.barh(categories, values5, left=values4+values3+values2+values1,
         color='tomato',label='Staten Island')

# Add legend, labels, and title
plt.legend()
plt.xlabel('Boroughs')
plt.ylabel('Population')
plt.title('Population in Five Boroughs of NYC')
plt.legend(bbox_to_anchor=(1.02, 0.8), loc='upper left', borderaxespad=0)
# Show plot
plt.savefig("nyc_pop_h.jpg",bbox_inches="tight")
plt.show()
```

```python
# 5 boroughs

plt.figure(figsize=(6, 4), dpi=600)
plt.scatter(categories,values1,label="Bronx",color='royalblue')
plt.plot(categories,values1,color='royalblue')
plt.scatter(categories,values2,label="Brooklyn",color='olive')
plt.plot(categories,values2,color='olive')
plt.scatter(categories,values3,label="Manhattan",color='gray')
plt.plot(categories,values3,color='gray')
plt.scatter(categories,values4,label="Queens",color='orange')
plt.plot(categories,values4,color='orange')
plt.scatter(categories,values5,label="Staten Island",color='tomato')
plt.plot(categories,values5,color='tomato')
#plt.scatter(categories,values1+values2,label="NYC")

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population in NYC')
plt.legend(bbox_to_anchor=(1.02, 0.8), loc='upper left', borderaxespad=0.5)

plt.savefig("nyc_pop_line.jpg",bbox_inches="tight")
plt.show()
```

## Percent Change 

```python
# change

change=pd.DataFrame({'1960-1950':df.iloc[1]-df.iloc[0],'1970-1960':df.iloc[2]-df.iloc[1],
                    '1980-1970':df.iloc[3]-df.iloc[2],'1990-1980':df.iloc[4]-df.iloc[3],
                     '2000-1990':df.iloc[5]-df.iloc[4],'2010-2000':df.iloc[6]-df.iloc[5],
                     '2020-2010':df.iloc[7]-df.iloc[6]})
change.head()
```

```python
total_change=change.sum()

plt.figure(figsize=(6, 4), dpi=600)
plt.stem(change.columns,total_change,linefmt='--g', basefmt='--k')
plt.stem(['1960-1950'],total_change[0:1],linefmt='--r', markerfmt='or')
plt.stem(['1980-1970'],total_change[2:3],linefmt='--r', markerfmt='or')
plt.xticks(rotation=45) 
plt.xlabel('')
plt.ylabel('')
plt.title('Population Change in NYC')
plt.savefig("nyc_pop_change.jpg",bbox_inches="tight")
plt.show()
```
